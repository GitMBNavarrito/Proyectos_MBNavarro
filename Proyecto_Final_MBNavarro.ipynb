# ANTECEDENTES

El servicio de telefonía CallMeMaybe busca mejorar su eficiencia y calidad. Para lograrlo, es crucial identificar a los operadores cuyo desempeño no cumple con los estándares de la empresa. La ineficacia se define a través de un conjunto de métricas clave que reflejan fallas en la gestión de llamadas.

**Métricas clave de rendimiento**

1. **Llamadas perdidas (internas y externas)**: Un alto número de llamadas sin atender por un operador indica falta de disponibilidad, una gestión ineficiente del volumen de trabajo o que el operador invierte demasiado tiempo en cada llamada. Esto impacta directamente la satisfacción del cliente.

2. **Tiempo de espera prolongado**: Un tiempo de espera excesivo para los clientes es una señal clara de ineficacia. Afecta negativamente la experiencia del usuario y la reputación del servicio.

3. **Número de llamadas salientes**: Un bajo volumen de llamadas salientes, cuando es parte de las responsabilidades del operador, puede ser un indicador de falta de proactividad. Sin embargo, esta métrica debe ser analizada junto con el tiempo invertido por llamada y los resultados obtenidos para comparar la calidad frente a la cantidad y el impacto real en el negocio.

**Factores externos y consideraciones adicionales** 

Para definir con precisión a un operador ineficiente, es fundamental considerar factores externos y contextuales que pueden influir en las métricas de rendimiento. Estos incluyen:

1. **Capacitación del operador**: La formación recibida por el operador para manejar el flujo y la complejidad de las llamadas.

2. **Eventos especiales**: Campañas o promociones de la empresa que puedan generar un flujo de llamadas inusualmente alto.

3. **Complejidad de las llamadas**: La falta de categorización de las llamadas puede distorsionar los datos, ya que algunas consultas son más complejas y requieren más tiempo que otras.

4. **Estacionalidad y horario**: El rendimiento puede variar significativamente según la hora del día o la temporada, por lo que los indicadores de rendimiento deben ajustarse a estas variables.

Al tomar en cuenta estas consideraciones, podremos establecer indicadores de rendimiento más justos y precisos, asegurando que la evaluación de los operadores sea un reflejo real de su desempeño y no de circunstancias ajenas a su control.

## 5.1 DEFINICIÓN DE LA HIPÓTESIS

Realizar pruebas de hipótesis para determinar si existe un grupo de operadores que presenta un **desempeño significativamente inferior**, al promedio de la compañía.

**Hipótesis 1** - Los operadores que tienen llamadas perdidas tienen más duración de llamadas.
- H0: La duración promedio de las llamadas de los operadores con llamadas perdidas es igual a la duración promedio de las llamadas de los operadores sin llamadas perdidas.
- Ha: Los operadores que tienen llamadas perdidas tienen una duración promedio de llamadas significativamente mayor que la de los operadores sin llamadas perdidas.

**Hipótesis 2** - No hay diferencia en el tiempo de espera de los clientes de alto valor vs. los demás; todos son tratados por igual.
- H0: El tiempo de espera promedio de los clientes de alto valor es igual al tiempo de espera promedio de los demás clientes.
- Ha: El tiempo de espera promedio de los clientes de alto valor es significativamente diferente al tiempo de espera promedio de los demás clientes.

**Hipótesis 3** - Los operadores que tienen llamadas salientes poseen el mayor número de llamadas perdidas.
- H0: El número de llamadas perdidas de los operadores con llamadas salientes es igual o menor que el número de llamadas perdidas de los operadores que no las realizan.
- Ha: El número de llamadas perdidas de los operadores con llamadas salientes es significativamente mayor que el número de llamadas perdidas de los operadores que no las realizan.

## 5.2 APLICACIÓN DE LA PRUEBA DE HIPÓTESIS

**Hipótesis 1**
Los operadores que tienen llamadas perdidas tienen una duración promedio de llamadas significativamente mayor que la de los operadores sin llamadas perdidas.

**Prueba de hipótesis**: Igualdad de las medias para muestras independientes:

- El grupo de operadores que ha tenido llamadas perdidas.

- El grupo de operadores que no ha tenido llamadas perdidas.

```python
df_llamadas_no_perdidas = operaciones[(operaciones['is_missed_call'] == False) & (operaciones['direction'] == "in")]
```

```python
operadores_no_perdidas = df_llamadas_no_perdidas.groupby('operator_id')['call_duration'].mean().round(2).reset_index()
operadores_no_perdidas.head()
```

```python
operadores_tiempo_and_perdidas = operadores_no_perdidas.merge(llamada_perdida_por_operador, on='operator_id', how='left')
operadores_tiempo_and_perdidas['tiene_perdida'] = operadores_tiempo_and_perdidas['calls_count'].apply(lambda x: 'no' if pd.isna(x) else 'si')
operadores_tiempo_and_perdidas.sample(5)
```

```python
df_con_perdida = operadores_tiempo_and_perdidas[operadores_tiempo_and_perdidas['tiene_perdida'] == 'si']
df_sin_perdida = operadores_tiempo_and_perdidas[operadores_tiempo_and_perdidas['tiene_perdida'] == 'no']

duracion_con_perdida = df_con_perdida['call_duration']
duracion_sin_perdida = df_sin_perdida['call_duration']

resultado_prueba_t = st.ttest_ind(duracion_sin_perdida, duracion_con_perdida, equal_var=False, alternative='less')

pvalue = resultado_prueba_t.pvalue
alpha = 0.05

print(f"\nValor p: {pvalue:.4f}")

if pvalue < alpha:
    print("Hay suficiente evidencia para rechazar la hipótesis nula. Los operadores con llamadas perdidas tienen una duración promedio de llamadas significativamente mayor.")
else:
    print("No hay suficiente evidencia para rechazar la hipótesis nula. La duración promedio de las llamadas es similar en ambos grupos.")
```

**Hipótesis 2**
Los operadores con pocas llamadas salientes tienen una duración promedio de la llamada significativamente mayor.

**Prueba de hipótesis**: Igualdad de las medias para muestras independientes:

- El grupo de operadores con pocas llamadas.

- El grupo de operadores con mayor número de llamadas.

```python
llamadas_out_with_tiempo = llamadas_out.groupby('operator_id').agg(cantidad_llamadas=('calls_count', 'sum'), promedio_llamada=('call_duration','mean')).reset_index()
```

```python
umbral = llamadas_out_with_tiempo['cantidad_llamadas'].quantile(0.25)


df_pocas_llamadas = llamadas_out_with_tiempo[llamadas_out_with_tiempo['cantidad_llamadas'] <= umbral]
df_muchas_llamadas = llamadas_out_with_tiempo[llamadas_out_with_tiempo['cantidad_llamadas'] > umbral]


resultado_prueba_t = st.ttest_ind(df_pocas_llamadas['promedio_llamada'],df_muchas_llamadas['promedio_llamada'], equal_var=False, alternative='greater')

pvalue = resultado_prueba_t.pvalue
alpha = 0.05

print(f"Valor p: {pvalue}")

if pvalue < alpha:
    print("Hay evidencia suficiente para rechazar la hipótesis nula. Los operadores con pocas llamadas salientes tienen una duración promedio significativamente mayor.")
else:
    print("No hay evidencia suficiente para rechazar la hipótesis nula. La duración promedio es similar en ambos grupos.")
```

**Hipótesis 3**
El número de llamadas perdidas de los operadores con llamadas salientes es significativamente mayor que el número de llamadas perdidas de los operadores que no las realizan.

**Prueba de hipótesis**: Igualdad de las medias para muestras independientes:

- Operadores que realizan llamadas salientes.

- Operadores que no realizan llamadas salientes.


```python
df_llamadas_no_perdidas_out = operaciones[(operaciones['is_missed_call'] == False) & (operaciones['direction'] == "out")]
```

```python
df_llamadas_no_perdidas_out_por_operador = df_llamadas_no_perdidas_out.groupby('operator_id')['calls_count'].sum().reset_index()
```

```python
operadores_out_and_perdidas = df_llamadas_no_perdidas_out_por_operador.merge(llamada_perdida_por_operador, on='operator_id', how='left')
operadores_out_and_perdidas['tiene_perdida'] = operadores_out_and_perdidas['calls_count_y'].apply(lambda x: 'no' if pd.isna(x) else 'si')
operadores_out_and_perdidas.sample(5)
```

```python
df_con_perdida1 = operadores_out_and_perdidas[operadores_out_and_perdidas['tiene_perdida'] == 'si']
df_sin_perdida1 = operadores_out_and_perdidas[operadores_out_and_perdidas['tiene_perdida'] == 'no']

out_con_perdida = df_con_perdida1['calls_count_x']
out_sin_perdida = df_sin_perdida1['calls_count_x']

resultado_prueba_t = st.ttest_ind(out_sin_perdida, out_con_perdida, equal_var=False, alternative='less')

pvalue = resultado_prueba_t.pvalue
alpha = 0.05

print(f"\nValor p: {pvalue:.4f}")

if pvalue < alpha:
    print("Hay suficiente evidencia para rechazar la hipótesis nula. El número de llamadas de los operadores que tienen llamadas perdidas es igual o menor que el número de llamadas de los operadores que no las realizan.")
else:
    print("No hay suficiente evidencia para rechazar la hipótesis nula. Los operadores que no tienen llamdas perdidas, tienen más llamdas salientes.")
```

## 5.3 RESULTADOS Y CONCLUSIONES DE LAS PRUEBAS

**Hipótesis 1** Los operadores que tienen llamadas perdidas tienen una duración promedio de llamadas significativamente mayor que la de los operadores sin llamadas perdidas. Valor p: 0.0000
Hay suficiente evidencia para rechazar la hipótesis nula. Los operadores con llamadas perdidas tienen una duración promedio de llamadas significativamente mayor.

La duración de las llamadas está directamente relacionada con la pérdida de llamadas. El valor p de 0.0000 es extremadamente bajo, lo que confirma que los operadores que manejan llamadas de mayor duración son significativamente más propensos a tener llamadas perdidas. Esto podría deberse a que están dedicando demasiado tiempo a casos complejos, dejando a otros clientes en espera hasta que cuelgan.

**Hipótesis 2** Los operadores con pocas llamadas salientes tienen una duración promedio de la llamada significativamente mayor. Valor p: 1.0
No hay evidencia suficiente para rechazar la hipótesis nula. La duración promedio es similar en ambos grupos.

Tener pocas llamadas salientes no significa ineficiencia. El valor p de 1.0 es muy alto, lo que indica que no hay una diferencia significativa en la duración promedio de las llamadas entre los operadores que hacen pocas llamadas salientes y los que hacen muchas. Esto apoya el hallazgo previo de que existe una especialización no declarada en el equipo. Algunos operadores se centran en llamadas entrantes, que pueden ser complejas, y esta es la razón de su bajo volumen de llamadas salientes.

**Hipótesis 3** El número de llamadas perdidas de los operadores con llamadas salientes es significativamente mayor que el número de llamadas perdidas de los operadores que no las realizan. Valor p: 0.0367
Hay suficiente evidencia para rechazar la hipótesis nula. El número de llamadas de los operadores que tienen llamadas perdidas es igual o menor que el número de llamadas de los operadores que no las realizan.

El volumen de llamadas salientes tiene un impacto directo en las llamadas perdidas. Con un valor p de 0.0367, se confirma que existe una relación. Los operadores que gestionan un alto volumen de llamadas salientes tienden a tener un número menor o igual de llamadas perdidas. Esto es contrario a lo que podría esperarse, ya que un operador ocupado con llamadas salientes no podría atender las entrantes.
